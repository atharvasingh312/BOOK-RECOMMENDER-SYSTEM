```python
ratingnum=ratings_with_bookname.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratingnum
```

```python
ratingnum.rename(columns={'Book-Rating':'num_rating'},inplace='True')
ratingnum
```

```python
ratingavg=ratings_with_bookname.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
ratingavg
```

```python
rating_df=ratingavg.merge(ratingnum,on="Book-Title")
rating_df
```

```python
rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace='True')
```

```python
final_rating=rating_df[rating_df['num_rating']>200]
```

```python
final_rating
```

```python
final_rating=final_rating.sort_values('avg_rating',ascending=0).head(100)
```

```python
#we show top 100 movies out of it
final_rating
```

```python
#since many books have different isbn we drop duplicates
final_rating=final_rating.merge(books,on='Book-Title')
final_rating=final_rating.drop_duplicates('Book-Title')[['Book-Title','avg_rating','num_rating','Book-Author','Year-Of-Publication','Image-URL-M']]
final_rating
```

# collaborative filtering approach

```python
ratings_with_bookname
```

```python
#index of experienced user
user=ratings_with_bookname.groupby('User-ID').count()['Book-Rating']>200
#boolean masking
exp_user=user[user].index
exp_user
```

```python
filter_rating=ratings_with_bookname[ratings_with_bookname['User-ID'].isin(exp_user)]
```

```python
filter_rating
```

```python
y=filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
book=y[y].index
```

```python
final=filter_rating[filter_rating['Book-Title'].isin(book)]
final
```

```python
pt=final.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
print(similarity_score.shape)
```

```python
def recommend(name):
    #fetching index
    index=np.where(pt.index==name)[0][0]
    distance=list(enumerate(similarity_score[index]))
    #reverse sorting
    distance=sorted(distance,key=lambda x:x[1],reverse=True)
    suggestion=distance[0:6]
    data=[]
    for i in suggestion:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1st to Die: A Novel')
```

```python
recommend("Outlander")
```

```python
recommend("Congo")
```
